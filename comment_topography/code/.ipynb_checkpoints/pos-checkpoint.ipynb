```python
def analyze_text_sentiment(text):
    client = language.LanguageServiceClient.from_service_account_json('../../../../scott-davis-remote-nlp-client.json')
    document = language.Document(content=text, type_=language.Document.Type.PLAIN_TEXT)

    response = client.analyze_sentiment(document=document)

    sentiment = response.document_sentiment
    res = [round(sentiment.score,3),round(sentiment.magnitude,3)]
    return(res)

```

```python

ivals = np.array(meta['GROUP'])
empty_list = []
for i in ivals:
    print(i)
    tdt = dt_unique_group[dt_unique_group['GROUP']==i]
    v = i
    if any(tdt.tolarge==True):
        [v,'NA','NA']
    if all(tdt.tolarge==False):
        tx = tdt['Letter.Text'].iloc[0]
        result = analyze_text_sentiment(tx)
        empty_list.append([v,result[0],result[1]])

    #dt['Letter.Text.Clean'][dt['UQID']==i]
proj = {'GROUP': [e[0] for e in empty_list], 'score':[e[1] for e in empty_list],'magnitude':[e[2] for e in empty_list]}
df = pd.DataFrame(data=proj)
```
