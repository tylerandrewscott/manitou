```python
from pydrive.auth import GoogleAuth
from pydrive.drive import GoogleDrive
#from google.colab import auth
from oauth2client.client import GoogleCredentials
import os
import pandas as pd
import numpy as np
import glob
gauth = GoogleAuth()
gauth.LocalWebserverAuth()
```

```python
import pandas as pd
drive = GoogleDrive(gauth)
pals = 'https://docs.google.com/spreadsheets/d/e/2PACX-1vTPHD71fQvTZpNljqRNVI6lD4b26wW86ptP_dy3R_qWaxkXyOE2QCuyTzroL_38mw/pub?output=csv'
pals_pd = pd.read_csv(pals)
pals_pd = pals_pd[pals_pd['DECISION TYPE']=='ROD']
pals_pd = pals_pd[pals_pd['PROJECT STATUS']=='Complete']
import re 
pals_pd.columns = [re.sub('– ','',x) for x in pals_pd.columns]
pals_pd = pals_pd[pals_pd[{'PROJECT NUMBER','PROJECT NAME','PROJECT CREATED','LMU FOREST'}].duplicated() == False]

meta_files = [f for f in drive.ListFile({'q': "'1m8HEJupt1nxMqAf35YyQ1-ykHY9W77Ou' in parents and trashed=false"})]
meta_titles = np.array([t['title'] for t in meta_files[0]])
meta_create = pd.to_datetime(np.array([t['createdDate'] for t in meta_files[0]]))

#read in project detail csv
pattern = re.compile('project_detail')
detail_versions = np.array([pattern.search(x)!=None for x in meta_titles])
detail_file = meta_files[0][np.where(meta_create == meta_create[detail_versions].max())[0][0]]
detail_file.GetContentFile('temp_project_detail.csv')
usfs_project_detail = pd.read_csv('temp_project_detail.csv')


#read in project overview csv
pattern = re.compile('project_overview')
detail_versions = np.array([pattern.search(x)!=None for x in meta_titles])
detail_file = meta_files[0][np.where(meta_create == meta_create[detail_versions].max())[0][0]]
detail_file.GetContentFile('temp_project_overview.csv')
usfs_project_overview = pd.read_csv('temp_project_overview.csv')
usfs_project_overview = usfs_project_overview[usfs_project_overview.duplicated()==False]

#read in document csv
pattern = re.compile('service_document')
detail_versions = np.array([pattern.search(x)!=None for x in meta_titles])
detail_file = meta_files[0][np.where(meta_create == meta_create[detail_versions].max())[0][0]]
detail_file.GetContentFile('temp_document.csv')
usfs_project_documents = pd.read_csv('temp_document.csv')
usfs_project_documents = usfs_project_documents[usfs_project_documents.duplicated()==False]
usfs_project_documents=usfs_project_documents.rename({'Proj_Num':'Project_Num'})
usfs_project_overview.rename(columns = {'Page':'Project_Page'},inplace=True)
#merge overview and details
usfs_projects = usfs_project_overview.merge(usfs_project_detail)
#filter
in_pals = np.isin(usfs_projects['Project_Num'],pals_pd['PROJECT NUMBER'])
usfs_projects = usfs_projects[in_pals]
in_projs = np.isin(usfs_project_documents['Project_Num'],usfs_projects['Project_Num'])
usfs_documents = usfs_project_documents[in_projs]
usfs_documents = usfs_documents[usfs_documents['Stage']!='Post-Decision']
```

```python
usfs_documents = usfs_documents[usfs_documents.Document_Name.str.contains(' Map|^Map|Map$') == False]
usfs_documents = usfs_documents[usfs_documents.Document_Name.str.contains('Comment|News Release|Presentation|Public Mtg|Public Meeting') == False]
usfs_documents = usfs_documents[usfs_documents['Stage']!='Scoping']
usfs_documents = usfs_documents[usfs_documents['Stage']!='Supporting']
```
